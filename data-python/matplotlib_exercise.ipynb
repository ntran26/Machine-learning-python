```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
%matplotlib inline
```

### Exercise 1

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

### Exercise 2

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y,'red')
ax2.plot(x,y,'red')

ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

### Exercise 3

```python
fig = plt.figure()

ax2 = fig.add_axes([0,0,1,1])
ax1 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xlim([20,22])
ax1.set_ylim([30,50])
ax1.set_title('zoom')

ax2.plot(x,z)
ax2.set_xlabel('x')
ax2.set_ylabel('z')
```

### Exercise 4

```python
fig, ax = plt.subplots(1,2)

ax[0].plot(x,y,'blue',lw=3,ls='--')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')

ax[1].plot(x,z,'red',lw=5,ls='-')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Z')

plt.tight_layout()
```
