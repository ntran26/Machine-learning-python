```python
import numpy as np
import pandas as pd
from numpy.random import randn

np.random.seed(101)
```

```python
# Create a table/data frame
df = pd.DataFrame(randn(5,4),['A','B','C','D','E'],['W','X','Y','Z'])
df
```

```python
# Conditional selection
df > 0
```

Select only TRUE rows of a selected column

```python
df[df['W'] > 0]
```

```python
df[df['Z'] < 0]
```

Conditional Selection

```python
bool = df['X']>0
result = df[bool]
result = result[['Y','Z']]
result
```

```python
df[df['X']>0][['Y','Z']]
```

Multiple conditions (use symbols: & for AND and | for OR)

```python
df[(df['W']>0) & (df['Y']>1)]
```

```python
df[(df['W']>0) | (df['Y']>1)]
```

Reset and set index

```python
df.reset_index(inplace=False)
```

```python
df['States'] = 'CA NY WY OR CO'.split()
df
```

```python
df.set_index('States',inplace=False)
```
